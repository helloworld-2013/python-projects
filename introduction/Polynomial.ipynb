```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(2)
a = np.random.normal(3.0, 1.0, 1000)
b = np.random.normal(50.0, 10.0, 1000) / a

plt.scatter(a, b)
plt.show()
```

```python
x = np.array(a)
y = np.array(b)
p4 = np.poly1d(np.polyfit(x,y,4))

xp = np.linspace(0,7,100)
plt.scatter(x,y)
plt.plot(xp,p4(xp),c='r')
plt.show()
```

```python
from sklearn.metrics import r2_score

# r-squared : 0 bad , 1 good , usually nearer 1 is better
print(r2_score(y, p4(x)))
```
